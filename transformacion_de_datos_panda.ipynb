#preparacion de datos

```python
import pandas as pd
```

##extraer tablas

```python
datos_host = pd.read_json("/content/datos_hosting.json")
```

```python
datos_host.head()
```

```python
datos_host = pd.json_normalize(datos_host["info_inmuebles"])
```

```python
#lista de la columnas de la tabla original
#del index 3 en adelante estan anidados
columnas = list(datos_host.columns)
```

```python
#sacar los datos anidados de cxala columna
datos_host=datos_host.explode(columnas[3:])
#de los datos sacar cada datop anidado de la lista de columbnas
datos_host
#excepto en comodidades q es un diccionario
```

```python
#resetear los indices
datos_host.reset_index(inplace=True, drop=True)
datos_host.head()

```

##cambiar tipo de datos

```python
#los datos se guadaron como object y no como int o str
datos_host.info()
```

###datos numericos

####enteros int

```python
import numpy as np
```

```python
#datos_host[columnas_int]=datos_host[columnas_int].astype(np.int64)
columnas_int = ["max_hospedes", "cantidad_baños", "cantidad_cuartos", "cantidad_camas"]
for columna in columnas_int:
    #datos_host[:, i] = np.array([int(x) for x in datos_host[:,i]])

    datos_host[columna] = pd.to_numeric(datos_host[columna], errors='coerce').astype('Int64')
```

```python
datos_host.info()
#todos los datos de la columna deben ser del mismo tipo
```

####datos float64

```python

datos_host["evaluacion_general"]=datos_host["evaluacion_general"].astype(np.float64)
datos_host.info()
```

####sacar caracteres de un dato(dato $)

```python
columnas_peso = ["cuota_deposito"	,"cuota_limpieza",	"precio"]
for columna in columnas_peso:
  datos_host[columna] = datos_host[columna].apply(lambda x: x.replace("$","").replace(",","").strip())


datos_host.head()
```

```python
##applymap sirve para series, para no hacer el for
#datos_host[["cuota_limpieza","precio"]].applymap(lambda x: x.replace("$","").replace(",",""))
```

```python
#pasar a float
for columna in columnas_peso:
    #datos_host[:, i] = np.array([int(x) for x in datos_host[:,i]])

    datos_host[columna] = pd.to_numeric(datos_host[columna], errors='coerce').astype('float64')
```
