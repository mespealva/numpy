# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/inmigrantes_canada.csv")
```

```python
df.head()
```

```python
df.info()
df.shape
```

```python
df.set_index("Pais", inplace=True)
```

```python
years = list(map(str,range(1980,2014)))
years
#incluye el primero y no el ultimo
```

```python
colombia = df.loc["Colombia", years]
#arreglo de los años en solo colombia
colombia
```

```python
col_dic= {"Año":colombia.index.to_list(),
          "Inmigrantes":colombia.values.tolist()}
#crear diccionario años con el valor de los indic3s del arreglo y inmigrantes con los valopres del alrreglo
#lista de años y despues lista de numeros inmigrantes
```

```python
d_col =pd.DataFrame(col_dic)
d_col
#convertirlo en el dataframe
```

##grafic0os matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4)) #tamaño de la grafica
plt.plot(d_col["Año"],d_col["Inmigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010"])
plt.title("Inmigracion de calumbiamnos a canada por los años")
plt.xlabel("Año")
plt.ylabel("Inmigreantes")

plt.show() #solomuestra la grafica sin el valor del espacio de memoria y q se muestre
```
