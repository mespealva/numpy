## valores nulos

```python
df_dep.isnull().sum()
#cuantos nulos hay en cada columna
```

```python
#nulos son =0
df_dep=df_dep.fillna(0)
df_dep.isnull().sum()
```

##remover datos inconcistentes

```python
#muestra los datos q son 0 en valor y condominio
df_dep.query("Valor==0|Condominio==0")
#muestra el indice donde se encuentran en la tabla
df_dep.query("Valor==0|Condominio==0 ").index
```

```python
df_dep_index=df_dep.query("Valor==0|Condominio==0 ").index
df_dep.drop(df_dep_index,axis=0,inplace=True)
```

```python
df_dep.drop("Tipo",axis=1,inplace=True)
#eliminar la columna tipo de la tabla, porq todos son departamentos
```

##aplicar filtros

```python
#dptos con 1 habitacion y alquiler menos a 4200
seleccion1=df_dep["Habitaciones"]==1
seleccion2=df_dep["Valor"] <4200
```

```python
filtro1=(seleccion1)&(seleccion2)
```

```python
#datos apartamentos de 2 dorm o mas y alquiler de 10500 y maypor a 70m2
filtro2=(df_dep["Habitaciones"]>=2)&(df_dep["Valor"]<10500)&(df_dep["Area"]>70)
```

##exportar datos a cvs

```python
#data frame completo a cvs, el de los dptos
#eliminar columna del indice anterior
#separa con ; como el original
df_dep.to_csv("inmuebles_ml.cvs",index=False, sep=";")

pd.read_csv("inmuebles_ml.cvs")
```

```python
#guardar los filtros
filtro1.to_csv("inmuebles_ml_filtro1.cvs",index=False, sep=";")
filtro2.to_csv("inmuebles_ml_filtro2.cvs",index=False, sep=";")
```
