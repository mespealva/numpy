##banco de datos, base de datos

###importar extencion

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData,Table, inspect, text
```

### primero crear el motor

```python
engine = create_engine("sqlite:///:memory:")
```

###importar

```python
archivo = "/content/clientes_banco.csv"
banco = pd.read_csv(archivo)
```

```python
banco.head()
```

```python
#de df a sql(nombre de la tabla, en q engine y no poner el index como columna)
banco.to_sql("cliente", engine, index=False)
```

```python
#revisar
insp = inspect(engine)
print(insp.get_table_names())
```

###consultas sql, query

```python
query = "SELECT * FROM cliente WHERE Categoria_de_renta = 'Empleado'"
```

```python
#la varieble tiene q convertirse en texto parea q el metodo sql lomlea como instruccion y no como variable
#conectar el motor de donde esta la tabla
#guardar los resaultados
empleados = pd.read_sql(sql=text(query), con = engine.connect())
empleados
```

```python
#crear la tabla empleado
empleados.to_sql("empleado", engine , index=False)
```

```python
#read_sql es para leer la consulta, para leer la tabla entera se puede usar este
pd.read_sql_table("empleado", engine)
```

```python
#puedo poner la comnsulta de ciertas coluymnas o especificaciones (colm como arreglo)
pd.read_sql_table("empleado", engine, columns=["ID_Cliente"	,"Edad",	"Grado_estudio"	])
```

###actualizar datos

```python
query = "SELECT * FROM cliente"
pd.read_sql(sql=text(query), con=engine.connect())
```

```python
#eliminar datos

from sqlalchemy.exc import SQLAlchemyError
query = "DELETE FROM cliente  WHERE ID_Cliente=6840104"

#tratar de agarrar el error por si ocurre metodo try
#delete no es consulta es una accion por eso es ejecutare
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("registro borrado exitosamente")
```

```python
#actualizar
from sqlalchemy.exc import SQLAlchemyError
query = "UPDATE cliente SET Grado_estudio='Nivel superior' WHERE ID_Cliente=6840222"

#tratar de agarrar el error por si ocurre metodo try
#delete no es consulta es una accion por eso es ejecutare
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("registro actualizado exitosamente")
```
